## 3. Explorar a estrutura

```python
import os, glob, json
ROOT='/content/pt100'
print('=== arvore (2 niveis) ===')
for d,subs,files in os.walk(ROOT):
    depth=d[len(ROOT):].count(os.sep)
    if depth<=1:
        print('  '*depth, os.path.basename(d) or ROOT, f'({len(files)} ficheiros, {len(subs)} pastas)')

print('\n=== contagem por extensao ===')
from collections import Counter
ext=Counter(os.path.splitext(f)[1].lower() for f in glob.glob(ROOT+'/**/*', recursive=True) if os.path.isfile(f))
for e,c in ext.most_common(15): print(f'  {e or "(sem)"}: {c}')

print('\n=== exemplos de ficheiros de anotacao (json/csv/txt) ===')
for pat in ('*.json','*.csv','*.txt'):
    hits=glob.glob(ROOT+'/**/'+pat, recursive=True)[:2]
    for h in hits:
        print('\n---', h, '---')
        try:
            with open(h) as f: head=f.read(1200)
            print(head)
        except Exception as e: print('erro a ler:', e)
```
